# thickness max pairwise plot

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_clipboard(header=None).transpose()
```

# 60X

```python
df_pos = df/2/9.26
df_neg = -df/2/9.26
```

# 40X

```python
df_pos = df/2/6.646
df_neg = -df/2/6.646
```

# Plot

```python
df.plot()
```

```python
for i in df.columns:
    plt.fill_between(df_pos.index, df_pos.ix[:, i], df_neg.ix[:, i], 
                     color='cornflowerblue', 
                     alpha = 0.1,
                    )
plt.ylim([-3, 3])
plt.axhline(y=-1)
plt.axhline(y=1)
plt.grid(color='lightgray', axis='y', alpha=0.5)
plt.axis('off')

plt.savefig('ok1489_fig_48h_max_pairwise.svg')
```
